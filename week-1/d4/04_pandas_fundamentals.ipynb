## Importing Pandas

Pandas is a Python library used for working with tabular data. It is commonly imported using the alias `pd`.

```python
import pandas as pd

print(pd.__version__)
```

## Pandas Series

A Series is a one-dimensional collection of values with an index. It is similar to one labeled column in a spreadsheet.

```python
scores = pd.Series([78, 85, 91, 88])
scores
```

### Series Data Type

A Series has a data type that describes the kind of values stored inside it. Pandas usually detects the data type automatically.

```python
integer_values = pd.Series([100, 102, 104])
decimal_values = pd.Series([100.0, 102.5, 104.2])
text_values = pd.Series(["A", "B", "C"])
boolean_values = pd.Series([True, False, True])

print("Integer dtype:", integer_values.dtype)
print("Decimal dtype:", decimal_values.dtype)
print("Text dtype:", text_values.dtype)
print("Boolean dtype:", boolean_values.dtype)
```

### Custom Index Labels

Pandas creates a numeric index by default, but custom labels can make the Series easier to understand and access.

```python
student_scores = pd.Series(
    [100, 85, 91, 66],
    index=["Ahmad", "Mohammad", "Omar", "Sami"]
)

student_scores
```

### Selecting by Label with loc

The `.loc` property selects a value using its index label.

```python
print(student_scores.loc["Ahmad"])
```

### Selecting by Position with iloc

The `.iloc` property selects a value using its integer position.

```python
print(student_scores.iloc[0])
```

### Comparing loc and iloc

Both `.loc` and `.iloc` can select values from a Series, but `.loc` uses labels while `.iloc` uses integer positions.

```python
print("Using loc:", student_scores.loc["Omar"])
print("Using iloc:", student_scores.iloc[2])
```

### Updating a Series Value

A value can be changed by selecting its label with `.loc` and assigning a new value.

```python
student_scores.loc["Omar"] = 94

student_scores
```

### Changing a Value Using Its Current Value

An existing value can also be increased or decreased instead of being completely replaced.

```python
student_scores.loc["Ahmad"] += 2
student_scores
```

### Filtering a Series

Filtering keeps only the values that match a condition. The condition first creates True and False values for every item in the Series.

```python
high_score_mask = student_scores >= 85

high_score_mask
```

```python
high_scores = student_scores[high_score_mask]

high_scores
```

### More Filtering Conditions

Different comparison operators can be used to filter a Series.

```python
scores_below_85 = student_scores[student_scores < 85]
scores_above_90 = student_scores[student_scores > 90]

print("Scores below 85:")
print(scores_below_85)

print("\nScores above 90:")
print(scores_above_90)
```

### Creating a Series from a Dictionary

A Series can also be created from a Python dictionary. The dictionary keys become the index labels, and the dictionary values become the Series values.

```python
study_hours_data = {
    "Sunday": 2,
    "Monday": 3,
    "Tuesday": 1.5,
    "Wednesday": 4
}

study_hours = pd.Series(study_hours_data)

study_hours
```

```python
print("Monday study hours:", study_hours.loc["Monday"])

long_study_days = study_hours[study_hours >= 3]

long_study_days
```

### Series Summary

In this section, I created Series objects from lists and dictionaries. I used custom index labels, selected values with `.loc` and `.iloc`, updated values, and filtered the Series using conditions.

## Pandas DataFrames

A DataFrame is a two-dimensional table with rows and named columns. It is similar to a spreadsheet and can store different types of data in different columns.

```python
employee_data = {
    "name": ["Ahmad", "Lina", "Omar"],
    "age": [21, 23, 25],
    "department": ["IT", "Design", "Finance"]
}

employees_df = pd.DataFrame(employee_data)

employees_df
```

### Rows, Columns, and Index

Each row represents one record, while each column represents one type of information. Pandas also creates an index for the rows.

```python
employees_df
```

### Custom Row Labels

A DataFrame uses numeric row labels by default, but custom labels can be provided when the DataFrame is created.

```python
employees_df = pd.DataFrame(
    employee_data,
    index=["employee_1", "employee_2", "employee_3"]
)

employees_df
```

### Selecting Rows with loc

The `.loc` property selects a row using its index label.

```python
employees_df.loc["employee_2"]
```

```python
print(type(employees_df))
print(type(employees_df.loc["employee_2"]))
```

### Selecting Rows with iloc

The `.iloc` property selects a row using its integer position.

```python
employees_df.iloc[0]
```

### Comparing loc and iloc

`.loc` selects by row label, while `.iloc` selects by integer position.

```python
print(employees_df.loc["employee_1"])
print("\n")
print(employees_df.iloc[0])
```

### Adding a New Column

A new column can be added by assigning a list of values to a new column name. The number of values must match the number of rows.

```python
employees_df["salary"] = [900, 1100, 1300]

employees_df
```

```python
employees_df["active"] = True

employees_df
```

### Adding a New Row

A new row can be created as a small DataFrame and then joined to the existing DataFrame using `pd.concat()`.

```python
new_employee=pd.DataFrame([{"name":"Rami",
                            "age": 19,
                             "department":"IT",
                              "salary":950,
                               "active":True}],index=["employee_4"])
print(new_employee)
print()
df=pd.concat([employees_df,new_employee])
print(df)
```

```python
new_employees = pd.DataFrame(
    [{"name": "Yousef","age": 24,"department": "IT","salary": 1200,"active": True},
     {"name": "Maya","age": 26,"department": "Finance","salary": 1400,"active": True}],
     index=["employee_5", "employee_6"]
)
employees_df = pd.concat([employees_df, new_employees])
```

### Checking the DataFrame

After adding columns and rows, I displayed the DataFrame again to make sure the new data was added correctly.

```python
employees_df
```

### Adding New Rows with concat

I created another dictionary for the new employees, converted it into a DataFrame, and combined it with the original DataFrame using `pd.concat()`.

```python
new_employee_data = {
    "name": ["Rami", "Sara"],
    "age": [19, 22],
    "department": ["IT", "Marketing"],
    "salary": [950, 1000],
    "active": [True, True]
}

new_employees_df = pd.DataFrame(
    new_employee_data,
    index=["employee_4", "employee_5"]
)

employees_df = pd.concat([employees_df, new_employees_df])

employees_df
```

## Filtering the Dataset

Filtering keeps only the rows that match a condition. I started with a simple condition, then combined two conditions to create a more specific subset.

```python
adult_passengers = titanic_df[titanic_df["Age"] >= 18]

adult_passengers[["Name", "Age", "Sex", "Pclass"]].head()
```

### Female Passengers Who Survived

I filtered the dataset to find female passengers who survived. I used `&` because both conditions must be true.

```python
female_survivors=titanic_df[(titanic_df["Sex"] == "female")&(titanic_df["Survived"] == 1)]
female_survivors[["Name", "Age", "Pclass", "Fare"]].head(10)
```

```python
female_survivor_count = female_survivors.shape[0]
print("Number of female survivors:", female_survivor_count)
```

The filtered result contains 233 female passengers who survived. They came from different ages and passenger classes.

## Aggregation

Aggregation functions summarize many values into one result. I used them to calculate simple statistics from numeric columns in the dataset.

```python
print("Average age:", titanic_df["Age"].mean())
print("Average fare:", titanic_df["Fare"].mean())
print("Minimum fare:", titanic_df["Fare"].min())
print("Maximum fare:", titanic_df["Fare"].max())
print("Number of known ages:", titanic_df["Age"].count())
```

## Grouping by Passenger Class

I grouped the passengers by ticket class. For each class, I calculated the average survival value, average fare, and number of passengers.

```python
group = titanic_df.groupby("Pclass")
print(group["Survived"].mean())
print()
print(group["Fare"].mean())
print()
print(group["PassengerId"].count())
```

### Groupby Result

The result shows that first-class passengers had the highest survival rate and paid the highest average fare. Third-class passengers had the lowest survival rate and also formed the largest passenger group.

## Data Cleaning

Before changing the dataset, I checked the number of missing values in each column.

```python
titanic_df.isna().sum()
```

```python
clean_titanic_df = titanic_df.copy()
age_median = clean_titanic_df["Age"].median()
clean_titanic_df["Age"] = clean_titanic_df["Age"].fillna(age_median)
most_common_port = clean_titanic_df["Embarked"].mode()[0]
clean_titanic_df["Embarked"] = clean_titanic_df["Embarked"].fillna(most_common_port)
clean_titanic_df = clean_titanic_df.drop(columns=["Cabin"])
clean_titanic_df.isna().sum()
```

### Missing Values Decision

I filled the missing `Age` values with the median because the median is less affected by unusually high or low values. I filled the missing `Embarked` values with the most common port because it is a categorical column. I removed the `Cabin` column because most of its values were missing.

### Duplicate Rows

I checked whether the dataset contained fully duplicated rows.

```python
duplicate_count = clean_titanic_df.duplicated().sum()
print("Duplicate rows:", duplicate_count)
clean_titanic_df = clean_titanic_df.drop_duplicates()
```

No fully duplicated rows were found, so removing duplicates did not change the dataset.
