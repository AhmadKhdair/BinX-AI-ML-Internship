# Day 5 — Matplotlib Fundamentals

## Introduction

Matplotlib is a Python library used to create data visualizations.

In this part of the notebook, I will learn how to create and customize the four main plot types:

- Line plots
- Scatter plots
- Bar plots
- Histograms

I will also learn how to combine multiple plots in one figure using subplots.

The goal is not only to display data, but also to choose the correct plot type and make each visualization clear and easy to understand.

## 1. Importing the Libraries

I will use NumPy to create and store numerical data, and Matplotlib to visualize it.

The `pyplot` module provides the main functions used to create figures, plots, labels, titles, and legends.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(10)
```

## 2. Creating Example Data

Before creating the plots, I will prepare a small example dataset representing the monthly performance of a fictional store.

The data includes:

- Monthly sales
- Monthly expenses
- Number of customers
- Monthly profit

Using a small dataset makes it easier to understand how each plot works before applying Matplotlib to a real dataset.

```python
months = np.array(["Jan", "Feb", "Mar", "Apr", "May", "Jun"])
sales = np.array([12000, 13500, 12800, 15000, 16200, 17500])
expenses = np.array([8000, 8500, 8300, 9200, 9700, 10400])
customers = np.array([240, 265, 250, 290, 315, 340])
profit = sales - expenses
```

```python
print("Months:", months)
print("Sales:", sales)
print("Expenses:", expenses)
print("Customers:", customers)
print("Profit:", profit)
```

## 3. Understanding the Structure of a Matplotlib Plot

A clear visualization usually contains:

- A figure size
- The plotted data
- A title
- An x-axis label
- A y-axis label
- A legend when multiple data series are shown
- A grid when it helps with reading values

The plot is displayed using `plt.show()`.

## 4. Line Plot

A line plot is used to show how a numeric value changes across an ordered or continuous sequence.

In this example, I will use a line plot to show how store sales changed from January to June.

```python
plt.figure(figsize=(10, 5))

plt.plot(months, sales)

plt.title("Monthly Store Sales")
plt.xlabel("Month")
plt.ylabel("Sales ($)")

plt.show()
```

### 4.1 Customizing the Line Plot

The basic line plot can be improved by adding markers, changing the line style, displaying a grid, and controlling the line width.

These customizations make individual data points and overall trends easier to identify.

```python
plt.figure(figsize=(10, 5))

plt.plot(
    months,
    sales,
    marker="o",
    linestyle="-",
    linewidth=2,
    label="Sales"
)

plt.title("Monthly Store Sales")
plt.xlabel("Month")
plt.ylabel("Sales ($)")

plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()

plt.show()
```

### 4.2 Comparing Multiple Lines

More than one line can be displayed in the same plot when the values share the same x-axis.

Here, monthly sales and expenses are shown together to compare how both changed over time.

```python
plt.figure(figsize=(10, 5))

plt.plot(
    months,
    sales,
    marker="o",
    linewidth=2,
    label="Sales"
)

plt.plot(
    months,
    expenses,
    marker="s",
    linewidth=2,
    label="Expenses"
)

plt.title("Monthly Sales and Expenses")
plt.xlabel("Month")
plt.ylabel("Amount ($)")

plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()

plt.show()
```

## 5. Scatter Plot

A scatter plot is used to examine the relationship between two numerical variables.

Each point represents one observation.

In this example, the plot compares the number of monthly customers with monthly sales to see whether higher customer counts are associated with higher sales.

```python
plt.figure(figsize=(9, 5))

plt.scatter(
    customers,
    sales,
    s=90,
    alpha=0.7
)

plt.title("Customers vs. Store Sales")
plt.xlabel("Number of Customers")
plt.ylabel("Sales ($)")

plt.grid(alpha=0.3)
plt.tight_layout()

plt.show()
```

### 5.1 Adding Labels to Scatter Points

Because the example contains only a small number of observations, each point can be labeled with its corresponding month.

This makes it easier to identify which month each point represents.

```python
plt.figure(figsize=(9, 5))

plt.scatter(
    customers,
    sales,
    s=90,
    alpha=0.7
)

for month, customer_count, sales_value in zip(months, customers, sales):
    plt.annotate(
        month,
        (customer_count, sales_value),
        textcoords="offset points",
        xytext=(5, 5)
    )

plt.title("Customers vs. Store Sales")
plt.xlabel("Number of Customers")
plt.ylabel("Sales ($)")

plt.grid(alpha=0.3)
plt.tight_layout()

plt.show()
```

## 6. Bar Plot

A bar plot is used to compare numerical values across categories.

In this example, each bar represents the store's profit for one month.

```python
plt.figure(figsize=(10, 5))
plt.bar(months, profit)
plt.title("Monthly Store Profit")
plt.xlabel("Month")
plt.ylabel("Profit ($)")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
```

## 7. Histogram

A histogram is used to examine the distribution of one numerical variable.

It divides numerical values into intervals called bins and counts how many observations fall inside each interval.

For this example, I will generate a larger sample of customer transaction values and examine their distribution.

```python
transaction_values = np.random.normal(
    loc=55,
    scale=12,
    size=200
)

transaction_values = np.round(transaction_values, 2)

transaction_values[:10]
```

```python
plt.figure(figsize=(9, 5))
plt.hist(
    transaction_values,
    bins=12,
    edgecolor="black"
)
plt.title("Distribution of Customer Transaction Values")
plt.xlabel("Transaction Value ($)")
plt.ylabel("Frequency")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
```

## 8. Creating Multiple Plots with Subplots

Subplots allow multiple visualizations to be placed inside one figure.

This is useful when several related plots need to be compared together.

The following figure contains:

1. A line plot for monthly sales and expenses
2. A scatter plot for customers and sales
3. A bar plot for monthly profit
4. A histogram for transaction values

```python
fig, axes = plt.subplots(
    2,
    2,
    figsize=(13, 9)
)
# Line plot
axes[0, 0].plot(
    months,
    sales,
    marker="o",
    label="Sales"
)
axes[0, 0].plot(
    months,
    expenses,
    marker="s",
    label="Expenses"
)
axes[0, 0].set_title("Monthly Sales and Expenses")
axes[0, 0].set_xlabel("Month")
axes[0, 0].set_ylabel("Amount ($)")
axes[0, 0].legend()
# Scatter plot
axes[0, 1].scatter(
    customers,
    sales,
    s=80,
    alpha=0.7
)
axes[0, 1].set_title("Customers vs. Sales")
axes[0, 1].set_xlabel("Number of Customers")
axes[0, 1].set_ylabel("Sales ($)")
# Bar plot
axes[1, 0].bar(
    months,
    profit
)
axes[1, 0].set_title("Monthly Profit")
axes[1, 0].set_xlabel("Month")
axes[1, 0].set_ylabel("Profit ($)")
# Histogram
axes[1, 1].hist(
    transaction_values,
    bins=12,
    edgecolor="black"
)
axes[1, 1].set_title("Transaction Value Distribution")
axes[1, 1].set_xlabel("Transaction Value ($)")
axes[1, 1].set_ylabel("Frequency")
for ax in axes.flat:
    ax.grid(alpha=0.3)
fig.suptitle(
    "Store Performance Dashboard",
    fontsize=16
)
plt.tight_layout(
    rect=[0, 0, 1, 0.96]
)
plt.show()
```

## 9. Pyplot and Object-Oriented Approaches

Matplotlib supports two common plotting approaches.

The `pyplot` approach is simple and convenient for individual plots.

The object-oriented approach creates explicit Figure and Axes objects, which provides better control when working with larger or more complex figures.

The object-oriented approach is especially useful when creating subplots.

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(
    months,
    sales,
    marker="o",
    linewidth=2
)
ax.set_title("Monthly Store Sales")
ax.set_xlabel("Month")
ax.set_ylabel("Sales ($)")
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## 10. Saving a Figure

Matplotlib figures can be saved as image files for use in reports, presentations, or project documentation.

The figure should be saved before calling `plt.show()`.

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(
    months,
    sales,
    marker="o",
    linewidth=2
)
ax.set_title("Monthly Store Sales")
ax.set_xlabel("Month")
ax.set_ylabel("Sales ($)")
ax.grid(alpha=0.3)
plt.tight_layout()
fig.savefig(
    "monthly_store_sales.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```

## 11. Choosing the Correct Plot Type

Different plot types answer different questions.

| Plot Type | Matplotlib Function | Main Purpose |
|---|---|---|
| Line plot | `plt.plot()` | Show a trend or change across an ordered sequence |
| Scatter plot | `plt.scatter()` | Examine the relationship between two numerical variables |
| Bar plot | `plt.bar()` | Compare numerical values across categories |
| Histogram | `plt.hist()` | Examine the distribution of one numerical variable |
| Subplots | `plt.subplots()` | Display multiple related plots in one figure |

Choosing the correct visualization is more important than adding unnecessary decoration.

## 12. Common Visualization Mistakes

Some common mistakes can make a visualization difficult to understand:

- Using the wrong plot type for the question
- Creating a plot without a title
- Leaving the axes unlabeled
- Displaying multiple lines without a legend
- Using too many histogram bins or too few bins
- Adding too many labels to a crowded scatter plot
- Creating subplots without enough figure space
- Using visual decoration that does not help explain the data

A good visualization should be clear, readable, and connected to a specific analytical question.
