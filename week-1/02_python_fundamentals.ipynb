## 2. Python Collections

Collections allow us to store multiple values in one variable.



### Lists

A list stores multiple values in a specific order. Lists are mutable, which means their items can be added, removed, or changed. Lists can also contain repeated values.

#### Accessing and Slicing

I created a list of scores and accessed its items using indexes. Python starts indexing from `0`, while `-1` can be used to access the last item. I also used slicing to select part of the list.

````python
scores=[73,86,86,95,76]
print(scores)
print(f"first score :{scores[0]}")
print(f"last score :{scores[-1]}")
print(f"second score :{scores[1]}")

print(f"scores from index 1 to 3  :{scores[1:4]}")
print(f"first three scores :{scores[:3]}")
    
````

#### Changing and Adding Items

Lists can be changed after they are created. In this part, I update one score, add one item to the end, insert an item at a specific position, and add multiple items to the list.

````python
scores[0]=75
scores.append(90)
scores.insert(1,80)
scores.extend([88,91])
print(scores)
````

#### Removing Items

The `remove()` method removes an item by its value. The `pop()` method removes an item by its position and also returns the removed value. Without an index, `pop()` removes the last item.

````python
scores.remove(86)
removedScore=scores.pop()
print(removedScore)
print(scores)
````

#### List Calculations

Python provides useful functions for working with numeric lists. Here, I calculate the number of scores, the highest and lowest values, the total, and the average.

````python
print(f"number of scores:{len(scores)}")
print(f"highest score:{max(scores)}")
print(f"lowest score:{min(scores)}")
print(f"total of scores :{sum(scores)}")
print(f"average :{sum(scores)/float(len(scores))}")
````

#### Searching in a List

The `count()` method shows how many times a value appears, while `index()` returns the position of its first occurrence. The `in` operator checks whether a value exists in the list.

````python
print(f"score 86 counts: {scores.count(86)}")
print(f"index of score 95 :{scores.index(95)}")
print(f"does score 100 exist: {100 in scores}")
print(f"does score 75 exist: {75 in scores}")
````

#### Sorting Lists

The `sorted()` function creates a new sorted list without changing the original one. I create both ascending and descending versions. I then sort the descending list again to confirm that it returns to ascending order.

````python
newScoresSorted=sorted(scores)
newScoresSortedDescending=sorted(scores,reverse=True)
print(scores)
print(newScoresSorted)
print(newScoresSortedDescending)
scoresSortedTwice=sorted(newScoresSortedDescending)
print(f"Ascending Sorted Scores :{newScoresSorted}")
print(f"scores sorted twice :{scoresSortedTwice}")
````

### Tuples

A tuple stores ordered values like a list, but its items cannot be changed after it is created. Tuples are useful for grouped values that should stay fixed, such as coordinates, dimensions, or information returned from a function.

#### Creating and Accessing a Tuple

In this part, I create a tuple containing course information. I then access individual items using positive and negative indexes and select part of the tuple using slicing.

````python
course_info = (
    "Python Fundamentals",
    2,
    8.0,
    True,
    "Python Fundamentals"
)

print("Course information:", course_info)
print("First item:", course_info[0])
print("Last item:", course_info[-1])
print("Items from index 1 to 3:", course_info[1:4])
````

#### Basic Tuple Operations

Tuples support several useful operations. I use `len()` to find the number of items, `count()` to count repeated values, `index()` to find a value's position, and `in` to check whether a value exists.

````python
print("Number of items:", len(course_info))

print(
    "Course name count:",
    course_info.count("Python Fundamentals")
)

print("Index of 8.0:", course_info.index(8.0))
print("Does True exist?", True in course_info)
````

#### Tuple Unpacking

Tuple unpacking allows each item in a tuple to be stored in a separate variable. The number of variables should match the number of items in the tuple.

````python
course_name, day_number, study_hours, is_completed, repeated_name = course_info

print("Course name:", course_name)
print("Day number:", day_number)
print("Study hours:", study_hours)
print("Completed:", is_completed)
print("Repeated name:", repeated_name)
````

#### Creating a One-Item Tuple

A tuple containing only one item must include a comma after the value. Without the comma, Python treats the value as a normal string, number, or other data type.

````python
company = ("BinX Tech",)

print("One-item tuple:", company)
print("Its type:", type(company))
````

#### Converting a Tuple to a List

Tuple items cannot be changed directly. When a value needs to be updated, the tuple can be converted into a list. The list can then be changed while the original tuple remains unchanged.

````python
course_info_list = list(course_info)

course_info_list[1] = 3

print("Original tuple:", course_info)
print("Changed list:", course_info_list)
````

### Sets

A set stores unique values and removes duplicates automatically. Unlike lists and tuples, sets do not use indexes and their item order is not guaranteed. Sets are useful for removing repeated values and comparing groups of data.

````python
languages = {"Python", "Java", "Python", "C++", "Java"}

print("Unique languages:", languages)
print("Number of unique languages:", len(languages))
print("Does Python exist?", "Python" in languages)

languages.add("SQL")
languages.update(["JavaScript", "R"])

print("After adding languages:", languages)

languages.remove("C++")
languages.discard("Ruby")

print("After removing languages:", languages)
````

## 3. Operators

Operators are symbols or words that perform calculations, compare values, and combine conditions.



### Arithmetic Operators

Arithmetic operators are used for mathematical calculations. Python supports normal division, floor division, remainders, and powers in addition to the basic operations.

````python
first_number = 17
second_number = 5

print("Addition:", first_number + second_number)
print("Subtraction:", first_number - second_number)
print("Multiplication:", first_number * second_number)
print("Division:", first_number / second_number)
print("Floor division:", first_number // second_number)
print("Remainder:", first_number % second_number)
print("Power:", first_number ** 2)
````

### Comparison Operators

Comparison operators compare two values and return either `True` or `False`. The `==` operator checks equality, while `=` is used to assign a value to a variable.

````python
score = 85
passing_score = 60

print("Equal to 85:", score == 85)
print("Not equal to 85:", score != 85)
print("Greater than passing score:", score > passing_score)
print("Less than passing score:", score < passing_score)
print("At least 85:", score >= 85)
print("At most 80:", score <= 80)
````

### Logical Operators

Logical operators combine conditions. `and` requires both conditions to be true, `or` requires at least one condition to be true, and `not` reverses a Boolean result.

````python
score = 85
attendance = 95
submitted_project = True

print(
    "Passed all requirements:",
    score >= 60 and attendance >= 90
)

print(
    "Has a high score or submitted the project:",
    score >= 90 or submitted_project
)

print("Project is not submitted:", not submitted_project)
````

### Membership Operators

The `in` and `not in` operators check whether a value exists inside a collection. When used with a dictionary, they check its keys.

````python
required_skills = ["Python", "Git", "Jupyter"]

print("Python is required:", "Python" in required_skills)
print("SQL is required:", "SQL" in required_skills)
print("Java is not required:", "Java" not in required_skills)

intern = {
    "name": "Ahmad",
    "completed_days": 2
}

print("Dictionary has a name key:", "name" in intern)
print("Dictionary has an email key:", "email" in intern)
````

## 4. Conditional Statements

Conditional statements allow a program to make decisions. Python uses `if` to check the first condition, `elif` to check additional conditions, and `else` when none of the previous conditions are true.


### Using `if` and `else`

In this example, the program checks whether a student reached the passing score.

````python
score = 72
passing_score = 60

if score >= passing_score:
    print("The student passed.")
else:
    print("The student failed.")
````

### Using `if`, `elif`, and `else`

Multiple conditions can be checked using `elif`. Python runs only the first block whose condition is true.

````python
score = 84

if score >= 90:
    result = "Excellent"
elif score >= 75:
    result = "Good"
elif score >= 60:
    result = "Passed"
else:
    result = "Failed"

print("Result:", result)
````

### Combining Conditions

Logical operators can combine conditions. In this example, the intern must have both a passing score and sufficient attendance.

````python
score = 82
attendance = 93

if score >= 60 and attendance >= 90:
    print("All requirements are completed.")
elif score >= 60:
    print("The score is enough, but attendance is too low.")
else:
    print("The score requirement is not completed.")
````

### Nested Conditions

A nested condition is an `if` statement inside another `if` statement. It is useful when the second decision only matters after the first condition is true.

````python
submitted_project = True
project_score = 88

if submitted_project:
    if project_score >= 60:
        print("The project was submitted and passed.")
    else:
        print("The project was submitted but did not pass.")
else:
    print("The project was not submitted.")
````

### Conditional Statements Practice

In this exercise, I check the intern's progress using the percentage of completed tasks, attendance, and project submission status.

````python
completed_tasks = 8
total_tasks = 10
attendance = 92
project_submitted = True

completion_percentage = completed_tasks / total_tasks * 100

print("Completion percentage:", completion_percentage)

if completion_percentage >= 80 and attendance >= 90 and project_submitted:
    print("Internship progress is excellent.")
elif completion_percentage >= 60 and attendance >= 80:
    print("Internship progress is acceptable.")
else:
    print("More work is needed.")

if not project_submitted:
    print("The project still needs to be submitted.")
````

## 5. Loops

Loops are used to repeat a block of code. A `for` loop is useful when working through a collection or a known sequence, while a `while` loop repeats as long as a condition remains true.

````python
skills = ["Python", "Git", "Jupyter"]

for skill in skills:
    print(skill)
````

### Using a `for` Loop

A `for` loop processes the items of a collection one at a time. In this example, each score is printed and added to a running total.

````python
scores = [75, 82, 90, 68]

total = 0

for score in scores:
    print("Current score:", score)
    total = total + score

print("Total:", total)
print("Average:", total / len(scores))
````

### Using `range()`

The `range()` function creates a sequence of numbers. It can define a starting value, a stopping value, and the amount added during each step.

````python
print("Numbers from 1 to 5:")

for number in range(1, 6):
    print(number)

print("Even numbers from 2 to 10:")

for number in range(2, 11, 2):
    print(number)
````

### Filtering Values with a Loop

A loop can check every value and store only the values that match a condition. Here, I create a new list containing only even numbers.

````python
numbers = [3, 8, 12, 7, 15, 20, 6]

even_numbers = []

for number in numbers:
    if number % 2 == 0:
        even_numbers.append(number)

print("Original numbers:", numbers)
print("Even numbers:", even_numbers)
````

### Using a `while` Loop

A `while` loop continues while its condition is true. The loop variable must be updated so the loop can eventually stop.

````python
completed_day = 1

while completed_day <= 5:
    print("Completed Day", completed_day)
    completed_day += 1

print("Week completed.")
````

### Stopping a Loop with `break`

The `break` statement ends a loop as soon as a specific condition is reached.

````python
scores = [75, 88, 91, 45, 80]

for score in scores:
    if score < 60:
        print("Low score found:", score)
        break

    print("Checked score:", score)
````

### Skipping an Item with `continue`

The `continue` statement skips the rest of the current repetition and moves to the next item.

````python
numbers = [5, -2, 8, -1, 10]

for number in numbers:
    if number < 0:
        continue

    print("Positive number:", number)
````

### Getting Indexes with `enumerate()`

The `enumerate()` function provides both the index and the value while looping through a collection.

````python
skills = ["Python", "Git", "Jupyter"]

for index, skill in enumerate(skills, start=1):
    print(index, "-", skill)
````

## 6. Functions

A function is a reusable block of code that performs a specific task. Functions make code easier to organize, test, and reuse. They can accept input through parameters and send a result back using `return`.

### Defining and Calling a Function

A function is defined using the `def` keyword. The code inside the function only runs when the function is called.

````python
def greet_intern():
    print("Welcome to Day 2 of the internship.")

greet_intern()
````

### Parameters and Arguments

Parameters are variables written in the function definition. Arguments are the actual values passed to the function when it is called.

````python
def greet_student(name):
    print(f"Hello {name}, welcome to the internship.")

greet_student("Ahmad")
greet_student("Mohammad")
````

### Returning a Value

The `return` statement sends a result back from a function. The returned value can be stored in a variable or used in another calculation.

````python
def calculate_average(numbers):
    average = sum(numbers) / len(numbers)
    return average

scores = [75, 80, 90, 85]

result = calculate_average(scores)

print("Average score:", result)
````

### Functions with Multiple Parameters

A function can receive more than one parameter. The arguments are normally matched to the parameters by their position.

````python
def calculate_completion(completed_tasks, total_tasks):
    percentage = completed_tasks / total_tasks * 100
    return percentage

completion = calculate_completion(8, 10)

print("Completion percentage:", completion)
````

### Default Parameters

A default parameter allows a function to use a predefined value when no argument is provided. The caller can still replace the default value when needed.

````python
def show_progress(completed_days, total_days=5):
    percentage = completed_days / total_days * 100
    return percentage

print("Default total days:", show_progress(3))
print("Custom total days:", show_progress(3, 10))
````

### Function Docstrings

A docstring is a short explanation written inside a function. It documents the purpose of the function and helps other developers understand how to use it.

````python
def calculate_total(price, quantity):
    """Calculate the total cost using a price and quantity."""
    total = price * quantity
    return total

print("Total cost:", calculate_total(25, 3))
help(calculate_total)
````

### Local Variables

A variable created inside a function is local to that function. It can be used while the function is running, but it is normally not available outside the function.

## 7. List Comprehensions

A list comprehension is a short way to create a new list from an existing collection. It can replace a simple loop that creates a list using `append()`.

### Creating a List with Two Methods

First, I create a list of squared numbers using a normal loop. Then I repeat the same operation using a list comprehension.

````python
numbers = [1, 2, 3, 4, 5]

squares_with_loop = []

for number in numbers:
    squares_with_loop.append(number ** 2)

squares_with_comprehension = [ number ** 2 for number in numbers ]

print("Using a loop:", squares_with_loop)
print("Using a list comprehension:", squares_with_comprehension)
````

### Filtering Even Numbers

### Choosing Between Two Values

An `if-else` expression inside a list comprehension can produce a different value for each item without removing any items.

````python
numbers = [1, 2, 3, 4, 5]

number_types = [
    "Even" if number % 2 == 0 else "Odd"for number in numbers
]

print("Numbers:", numbers)
print("Number types:", number_types)
````

## 8. Object-Oriented Programming

Object-oriented programming organizes related data and behavior inside classes. A class acts as a blueprint, while an object is a specific instance created from that class.

Data stored inside an object is called attributes, and functions defined inside a class are called methods.

### Creating a Class and Its Attributes

The `__init__` method initializes a new object. The `self` parameter refers to the object currently being created or used.

In this example, every student record stores a name and a score as object attributes.

````python
class StudentRecord:
    def __init__(self, name, score):
        self.name = name
        self.score = score
student_1 = StudentRecord("Ahmad", 88)
student_2 = StudentRecord("Sara", 95)
print(student_1.name)
print(student_2.name)
print(student_1.score)
print(student_2.score)
````

### Creating Objects

An object is created by calling the class and providing the required values. Each object stores its own attribute values, and the attributes can be accessed using dot notation.

````python
student_1 = StudentRecord("Ahmad", 88)
student_2 = StudentRecord("Sara", 95)

print("First student:", student_1.name)
print("First score:", student_1.score)

print("Second student:", student_2.name)
print("Second score:", student_2.score)
````

### Adding a Method

A method is a function defined inside a class. It can use `self` to access the attributes of the current object.

The `get_summary` method returns a sentence containing the student's name and score.

````python
class StudentRecord:
    def __init__(self, name, score):
        self.name = name
        self.score = score

    def get_summary(self):
        return f"{self.name} received a score of {self.score}."
````

````python
student_1 = StudentRecord("Ahmad", 88)
student_2 = StudentRecord("Sara", 95)

print(student_1.get_summary())
print(student_2.get_summary())
````

### Student Data Record Class

In this exercise, I define a class that represents a student data record. Each object stores the student's name and score as attributes.

The class contains methods that return a summary and check whether the student passed. This demonstrates how a class combines related data and behavior.

````python
class StudentRecord:
    """Represent a student record containing a name and score."""

    def __init__(self, name, score):
        self.name = name
        self.score = score

    def get_summary(self):
        return f"{self.name} received a score of {self.score}."

    def has_passed(self, passing_score=60):
        return self.score >= passing_score


student_1 = StudentRecord("Ahmad", 88)
student_2 = StudentRecord("Sara", 55)

print(student_1.get_summary())
print("Passed:", student_1.has_passed())

print(student_2.get_summary())
print("Passed:", student_2.has_passed())
````

### What I Learned

A class is a blueprint used to create objects. Each object can store its own data in attributes and perform related operations using methods.

The `__init__` method initializes a new object, while `self` refers to the specific object currently using a method. Multiple objects created from the same class remain independent from each other.
